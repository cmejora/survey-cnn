## Program to obtain the frequency distribution graph of the p01 question (What master's program are you enrolled in?) from the responses of the applied questionnaire

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
from scipy import signal
from scipy.stats import norm
import statistics
import math

# Data 
df = pd.read_csv('/Users/carlos/Documents/Programas_Python_GE/enc_datosnum.csv',delimiter=",",encoding = "ISO-8859-1")
data = pd.DataFrame(df)

varmue = data['P01']  

# Get frequency of each category
frecuencia_varmue = {}
for g in varmue:
    if g in frecuencia_varmue:
        frecuencia_varmue[g] += 1
    else:
        frecuencia_varmue[g] = 1

# Extract labels and frequency values
etiquetas = list(frecuencia_varmue.keys())
#print (etiquetas)
frecuencias = list(frecuencia_varmue.values())
#print(frecuencia_varmue)

# Create the bar chart
fig, ax = plt.subplots()
ax.grid(axis="y", linewidth=0.3)  # Only horizontal lines
ax.set_axisbelow(True)
plt.bar(etiquetas, frecuencias, color='#3eaaf6', linewidth=0.5, edgecolor="black", width=0.94)

x = np.linspace(0.0, len(etiquetas)+0.8, 1000)
media = np.mean(varmue)
desviacion_estandar = np.std(varmue,ddof=1)
y = stats.norm.pdf(x, media, desviacion_estandar)

# Scale the normal curve to fit the frequency scale
y = y * max(frecuencias) / max(y)


plt.plot(x, y, 'black', label= 'Normal curve', linewidth=2)

# Add the text inside the graph
texto = "Mean = "+str(round(media,2))+"\nσ = "+str(round(desviacion_estandar,3))+"\nN = "+str(112)

plt.text(len(etiquetas) + 1.1, 60., texto, fontsize=10, color='black')

# Add title and labels
plt.title('''What master's program are you enroll in?''')
plt.xlabel('''Relevance level''')
plt.ylabel('Frequency')

# Add legend
plt.legend()

# Display the graph
#plt.show()

#plt.show()
plt.savefig('HistP01')
```
