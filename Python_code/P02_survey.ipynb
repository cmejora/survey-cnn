## Descriptive statistics

```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis

df = dconcat

# Function to obtain descriptive statistics
def calcular_estadisticas(df):
    resultados = {}
    
    for columna in df.columns:
        datos = df[columna]
        if columna == 'ï»¿K':
            continue
       
        estadisticas = {
            'Mean': datos.mean(),
            'Median': datos.median(),
            'Mode': datos.mode()[0],  # There can be multiple modes; we take the first one
            'Standard Deviation': datos.std(),
            'Variance': datos.var(),
            'Skewness': skew(datos),
            'Standar error of Skewness': np.sqrt(6 / len(datos)),  # Standard error approximation of skewness
            'Kurtosis': kurtosis(datos),
            'Range': datos.max() - datos.min()
        }
        
        resultados[columna] = estadisticas
    
    return pd.DataFrame(resultados)

# Obtain the descriptive statistics
estadisticas_df = calcular_estadisticas(df)

# Dispay descriptive statistics
print(estadisticas_df)
```
